# Team Metric

```python
#we will be analyzing wickets, total score and average run-rate
df_filtered = df[(df['innings'] == 1) | (df['innings'] == 2)]
```

```python
# segregating the super overs from normal overs 
super_overs = df[(df['innings'] == 3) | (df['innings'] == 4) | (df['innings'] == 5) | (df['innings'] == 6)]
```

```python
indexes = [i for i in range(207881)]

df_filtered.index = indexes
```

```python
df_filtered.index
```

```python
# creating a separate dataframe for better analysis 
team_metric = pd.DataFrame(columns=['match_id','season','venue','batting_team','total_score','wickets','end_over','arr'])
```

```python
# function helps in determining completion of overs 
def whole_over(x):
    if str(x)[-1] == '6':
        x = int(x) + 1
    return x
```

```python
#this block of code is for inputting values in the new dataframe 
l1 = {}
l2 = {}
l3 = {}
l5 = {}
dummy = [i for i in range(1748)]
l6 = dict.fromkeys(dummy, 0)
l7 = dict.fromkeys(dummy, 0)
l8 = {}
l9 = {}
count = 0

for i in range(len(df_filtered['innings'])):
    if(i != 207880 and df_filtered.iloc[i, 4] == df_filtered.iloc[i + 1, 4]):
        l1[count] = int(df_filtered.iloc[i, 0])
        l2[count] = int(df_filtered.iloc[i, 1])
        l3[count] = df_filtered.iloc[i, 3]
        l5[count] = df_filtered.iloc[i, 6]
        l6[count] += df_filtered.iloc[i, 11] + df_filtered.iloc[i, 12]
        if pd.notna(df_filtered.iloc[i, 17]) and df_filtered.iloc[i, 4] == 1:
            l7[count + 1] += 1
            if count == 483:
                print(l7)
        elif pd.notna(df_filtered.iloc[i, 17]) and df_filtered.iloc[i, 4] == 2:
            l7[count - 1] += 1
            if count == 483:
                print(l7)
        
    else:
        l8[count] = df_filtered.iloc[i, 5]
        l8[count] = whole_over(l8[count])
        l9[count] = l6[count]/ l8[count]
        count += 1
```

```python
# appending all the extracted values in the new dataframe 
team_metric['match_id'] = list(l1.values())
team_metric['season'] = list(l2.values())
team_metric['venue'] = list(l3.values())
team_metric['batting_team'] = list(l5.values())
team_metric['total_score'] = list(l6.values())
team_metric['wickets'] = list(l7.values())
team_metric['end_over'] = list(l8.values())
team_metric['arr'] = list(l9.values())
```

```python
# observing the new obtained dataframe 
#arr stands for average runrates of individual teams. We can easily get insights of the run-rate from this dataframe 
#This new dataframe helps us analyze end over of a match, total runs scored and wickets lost by a team. 
team_metric
```
