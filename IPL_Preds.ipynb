```python
def whole_over(x):
    if str(x)[-1] == '6':
        x = int(x) + 1
    return x
```

```python
l1 = {}
l2={}
l3={}
l5={}
dummy = [i for i in range(1752)]
wickets_taken = df_filtered[df_filtered['wicket_type'].notna()].index
l6 = dict.fromkeys(dummy, 0)
l7 = dict.fromkeys(dummy, 0)
l8 = {}
l9 = {}
count = 0

for i in range(len(df_filtered['batting_team'])):
    if(i != 207881 and df_filtered.iloc[i, 6] == df_filtered.iloc[i+1, 6]):
        l1[count] = df_filtered['match_id'][i]
        l2[count]=df_filtered['season'][i]
        l3[count]=df_filtered['venue'][i]
        
        l5[count]=df_filtered['batting_team'][i]
        l6[count] = l6[count] + df_filtered['runs_off_bat'][i] + df_filtered['extras'][i]
        if i in wickets_taken:
            l7[count] += 1
    else:
        l8[count] = (df_filtered['ball'][i])
        l8[count] = whole_over(l8[count])
        l9[count] = l6[count]/ l8[count]
        count += 1
```

```python
team_metric['match_id'] = list(l1.values())
team_metric['season'] = list(l2.values())
team_metric['venue'] = list(l3.values())
team_metric['batting_team'] = list(l5.values())
team_metric['total_score'] = list(l6.values())
team_metric['wickets'] = list(l7.values())
team_metric['end_over'] = list(l8.values())
team_metric['arr'] = list(l9.values())
```
